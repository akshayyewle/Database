```python
%%sql 

SELECT id, home_team_goal, away_team_goal,
CASE WHEN home_team_goal > away_team_goal THEN 'Home Team Win'
    WHEN home_team_goal < away_team_goal THEN 'Away Team Win'
    ELSE 'Tie' END AS outcome
FROM match
WHERE season = '2013/2014';
```

# **Module 04: Window Function**

## **Window Function**

```python
%%sql 
-- How many goals were scored in each match in 2011/2012, and how did that compare to the average?

SELECT date,
    (home_team_goal + away_team_goal) AS goals,
    (SELECT AVG(home_team_goal + away_team_goal)
        FROM match
        WHERE season = '2011/2012') AS overall_avg
FROM match
WHERE season = '2011/2012';
```

## **Create RANK**

```python
%%sql
-- What is the rank of matches based on number of goals scored? 

SELECT date,
    (home_team_goal + away_team_goal) AS goals,
    RANK() OVER(ORDER BY home_team_goal + away_team_goal DESC) AS goals_rank
FROM match
WHERE season = '2011/2012';
```

## **Window Function With Partition**

### **Without Partioning**

```python
%%sql 
-- How many goals were scored in each match, and how did that compare to the overall average?

SELECT
date,
(home_team_goal + away_team_goal) AS goals,
AVG(home_team_goal + away_team_goal) OVER() AS overall_avg
FROM match;
```

### **With Partitioning**

```python
%%sql 
-- How many goals were scored in each match, and how did that compare to the season's average?

SELECT date,
    (home_team_goal + away_team_goal) AS goals,
    AVG(home_team_goal + away_team_goal) OVER(PARTITION BY season) AS season_avg
FROM match;
```

### **Partition By Multiple Cols**

```python
%%sql 

SELECT c.name, m.season,
    (home_team_goal + away_team_goal) AS goals,
    AVG(home_team_goal + away_team_goal)
    OVER(PARTITION BY m.season, c.name) AS season_ctry_avg
FROM country AS c
LEFT JOIN match AS m
ON c.id = m.country_id
```

## **Sliding Window Function**

```python
%%sql 

SELECT date, home_team_goal, away_team_goal, SUM(home_team_goal)
    OVER(ORDER BY date ROWS BETWEEN
        UNBOUNDED PRECEDING AND CURRENT ROW) AS running_total
FROM match
WHERE hometeam_id = 8456 AND season = '2011/2012';
```

# **Practice Queries**

```python
%%sql 

SELECT 
    -- Match Details
    m.id AS match_id,m.season, c.name AS country,

    -- Goal Details
    home_team_goal AS home, 
    away_team_goal AS away,
    (home_team_goal + away_team_goal) AS total_goal,

    -- Team Details
    # home_team_api_id AS home_team_id,
    t1.team_short_name as home_team,
    # away_team_api_id AS away_team_id,
    t2.team_short_name as away_team,

    -- Conditional Values Column 
    -- Match Result 
    CASE WHEN home_team_goal > away_team_goal THEN 'Home Team Win'
        WHEN home_team_goal < away_team_goal THEN 'Away Team Win'
        ELSE 'Draw' END AS match_result

FROM match AS m
INNER JOIN country AS c ON c.id = m.country_id
INNER JOIN teams AS t1 ON t1.team_api_id = m.home_team_api_id
INNER JOIN teams AS t2 ON t2.team_api_id = m.away_team_api_id

-- Filtering Criteria
# WHERE season = '2008/2009'

-- Sorting Criteria
ORDER BY total_goal DESC;
```
